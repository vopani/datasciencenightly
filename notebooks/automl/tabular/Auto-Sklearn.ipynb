## 5. Run binary classification
`AutoSklearnClassifier` can be used to run a classifier. Let's use the breast cancer dataset available in sklearn as an example.   
***P.S.*** *The runtime is limited to 60 seconds for faster training*

```python
from autosklearn.classification import AutoSklearnClassifier
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer(as_frame=True)['data']
target = load_breast_cancer(as_frame=True)['target']

data
```

```python
target
```

```python
sklearn_aml = AutoSklearnClassifier(time_left_for_this_task=60)
sklearn_aml.fit(X=data, y=target)
```

## 11. Check the run history
The history of all the runs are available in `automl_.runhistory_.data`.

```python
for i, run_key in enumerate(sklearn_aml.automl_.runhistory_.data):
    print(f'## Run {i+1} ##')
    print(run_key)
    print(sklearn_aml.automl_.runhistory_.data[run_key])
    print('')
```

## 13. Use validation split
A separate `validation dataset` can be provided during training. Let's split the dataset into train-valid using 80-20 split.

```python
from sklearn.model_selection import train_test_split

train, valid, target_train, target_valid = train_test_split(data, target, train_size=0.8, random_state=41)
print(f'Train Data shape: {train.shape}')
print(f'Valid Data shape: {valid.shape}')

sklearn_aml = AutoSklearnClassifier(time_left_for_this_task=60)
sklearn_aml.fit(X=train, y=target_train, X_test=valid, y_test=target_valid, dataset_name='breast_cancer')
print(sklearn_aml.sprint_statistics())
```

## 26. Run using SMAC optimizer
[Sequential Model-Based Optimization for General Algorithm Configuration](http://ml.informatik.uni-freiburg.de/papers/11-LION5-SMAC.pdf) (`SMAC`) is the default hyperparameter optimization algorithm used.

```python
sklearn_aml = AutoSklearnClassifier(time_left_for_this_task=60)
sklearn_aml.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml.sprint_statistics())
```

## 27. Run using ROAR optimizer
[Random Online Aggressive Racing](http://ml.informatik.uni-freiburg.de/papers/11-LION5-SMAC.pdf) (`ROAR`) can be used for hyperparameter optimization instead of the default SMAC.

```python
# https://automl.github.io/auto-sklearn/master/examples/60_search/example_random_search.html
from smac.facade.roar_facade import ROAR
from smac.scenario.scenario import Scenario

def get_roar_object_callback(
    scenario_dict,
    seed,
    ta,
    ta_kwargs,
    metalearning_configurations,
    n_jobs,
    dask_client,
):
    """Random online adaptive racing."""

    if n_jobs > 1 or (dask_client and len(dask_client.nthreads()) > 1):
        raise ValueError("Please make sure to guard the code invoking Auto-sklearn by "
                         "`if __name__ == '__main__'` and remove this exception.")

    scenario = Scenario(scenario_dict)
    return ROAR(
        scenario=scenario,
        rng=seed,
        tae_runner=ta,
        tae_runner_kwargs=ta_kwargs,
        run_id=seed,
        dask_client=dask_client,
        n_jobs=n_jobs,
    )

sklearn_aml = AutoSklearnClassifier(time_left_for_this_task=60, initial_configurations_via_metalearning=0,
                                    get_smac_object_callback=get_roar_object_callback)
sklearn_aml.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml.sprint_statistics())
```

## 28. Run using random search
`Random search` can also be used for hyperparameter optimization instead of the default SMAC.

```python
# https://automl.github.io/auto-sklearn/master/examples/60_search/example_random_search.html
from smac.facade.roar_facade import ROAR
from smac.scenario.scenario import Scenario

def get_random_search_object_callback(
        scenario_dict,
        seed,
        ta,
        ta_kwargs,
        metalearning_configurations,
        n_jobs,
        dask_client
):
    """Random search."""

    if n_jobs > 1 or (dask_client and len(dask_client.nthreads()) > 1):
        raise ValueError("Please make sure to guard the code invoking Auto-sklearn by "
                         "`if __name__ == '__main__'` and remove this exception.")

    scenario_dict['minR'] = len(scenario_dict['instances'])
    scenario_dict['initial_incumbent'] = 'RANDOM'
    scenario = Scenario(scenario_dict)
    return ROAR(
        scenario=scenario,
        rng=seed,
        tae_runner=ta,
        tae_runner_kwargs=ta_kwargs,
        run_id=seed,
        dask_client=dask_client,
        n_jobs=n_jobs,
    )

sklearn_aml = AutoSklearnClassifier(time_left_for_this_task=60, initial_configurations_via_metalearning=0,
                                    get_smac_object_callback=get_random_search_object_callback)
sklearn_aml.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml.sprint_statistics())
```

## 35. Make results reproducible
Models can be reproducible by setting a `seed` and limiting the training to a fixed number of runs.

```python
# without reproducibility
sklearn_aml_1 = AutoSklearnClassifier(time_left_for_this_task=300)
sklearn_aml_1.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml_1.sprint_statistics())

sklearn_aml_2 = AutoSklearnClassifier(time_left_for_this_task=300)
sklearn_aml_2.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml_2.sprint_statistics())
```

```python
# with reproducibility
sklearn_aml_1 = AutoSklearnClassifier(time_left_for_this_task=300, seed=41, initial_configurations_via_metalearning=0,
                                      disable_evaluator_output=False, smac_scenario_args={'runcount_limit': 3})
sklearn_aml_1.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml_1.sprint_statistics())

sklearn_aml_2 = AutoSklearnClassifier(time_left_for_this_task=300, seed=41, initial_configurations_via_metalearning=0,
                                      disable_evaluator_output=False, smac_scenario_args={'runcount_limit': 3})
sklearn_aml_2.fit(X=data, y=target, dataset_name='breast_cancer')
print(sklearn_aml_2.sprint_statistics())
```
